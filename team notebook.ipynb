# Market Segment Analysis

```python
d4 = pd.read_csv('Datasets\\Statewise Resource Sharing Hubs.csv')
d5 = pd.read_csv('Datasets\\Machinery_Requested_booked_Hired.csv')
```

```python
d4.head()
```

```python
d5.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Merging the datasets on common columns like 'State lgd code', 'Year', etc.
merged_data = pd.merge(d4, d5, on=['Country', 'State lgd code', 'State', 'Yearcode'], how='inner')
```

```python
merged_data.head()
```

```python
# Exploratory Data Analysis
merged_data.describe()
sns.heatmap(merged_data.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap="Blues")
```

```python
# Feature Engineering
merged_data['percent_implements_booked'] = (merged_data['Implements booked'] / merged_data['Implements requested']) * 100
```

```python
# Filling missing values
merged_data.fillna(0, inplace=True)
```

```python
merged_data.head()
```

```python
# Principal Component Analysis (PCA)
merged_data = merged_data.select_dtypes(include=['float64', 'int64'])
scaler = StandardScaler()
data_scaled = scaler.fit_transform(merged_data.iloc[:, 6:]) # Select the appropriate columns
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data_scaled)
```

```python
from sklearn.cluster import KMeans

# K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
merged_data['cluster'] = kmeans.fit_predict(pca_result)
```

```python
# Visualize the clusters
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=merged_data['cluster'], palette='Set1')
plt.title('FarmSwap Segmentation using K-Means')
```

```python
# Analyze each cluster
cluster_profile = merged_data.groupby('cluster').mean()
```

```python
# Profile key features in each segment
cluster_profile[['Custom hiring centers', 'Hi-tech, high productive equipment hub', 'percent_implements_booked']].plot(kind='bar', figsize=(10, 6))
plt.title("Cluster Profiles for FarmSwap Segments")
```
